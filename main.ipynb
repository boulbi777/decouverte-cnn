## Q1. La fonction buildCNN

Ici, nous définissons une fonction basique de création d'un réseau de neurones convolutif à partir des informations sur les dimensions des couches d'entrées et de sorties du réseau. Elle retourne un objet de type **keras.engine.training.Model** contenant le structure de notre réseau.

### 1. Importation de toutes les librairies nécessaires

```python
import os
import numpy as np
import math
from time import time
from datetime import datetime

#Librairies keras pour les fonctionalités de notre réseau
from keras.models import Sequential, Model
from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten, Dropout, Input
from keras.optimizers import Adam, rmsprop, sgd
from keras.callbacks import EarlyStopping, LearningRateScheduler, TensorBoard
import keras.backend as K
from keras.models import Model, load_model
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

### 2. Définition de la fonction

```python
def build_CNN(input_dim, output_dim, lr=0.001):
    
    # Initialisation du réseau
    model  = Sequential()
    
    #Différentes couches
    model.add(Conv2D(filters=30, kernel_size=(3,3), input_shape=input_dim))
    model.add(Activation(activation="relu"))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(rate=0.1))

    #-----------------------------------------------------------------------------------#

    model.add(Conv2D(filters=60, kernel_size=(3,3)))
    model.add(Activation(activation="relu"))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(rate=0.1))

    #----------------------------------------------------------------------------------#

    model.add(Conv2D(filters=120, kernel_size=(3,3)))
    model.add(Activation(activation="relu"))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(rate=0.1))

    #----------------------------------------------------------------------------------#

    model.add(Flatten())
    model.add(Dense(units=150))
    model.add(Activation(activation="relu"))
    model.add(Dropout(rate=0.3))

    #----------------------------------------------------------------------------------#

    model.add(Dense(units=output_dim))
    model.add(Activation(activation="softmax"))    
    
    # initialiser l'optimiseur RMSprop
    opt = rmsprop(learning_rate=lr, decay=1e-6)
    
    #Compilation du modèle
    model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    return model
```

### 3. A quoi sert le DropOut ?

Dans la définition de la fonction, plusieurs couches sont définies les unes après les autres. Sur chaque couche, on précise le type de convultion souhaité (2D dans notre cas), la fonction d'activation, etc...

Le **droupOut** dans un réseau de neurone sert essentiellement à éviter de l’overfitting lors de l'entraînement du réseau. En effet, dans les CNN, le passage plusieurs fois sur les mêmes pixels d'une image donnée (principe de la convolution) peut rapidement conduire à ce phénomène de surapprentissge où les neurones des couches cachées (ou intermédiaire) ajustent leurs poids qui se collent trop aux données d'entraînement. Le dropOut est donné avec un pourcentage qui détermine la proportion de neurones à tuer à la sortie de chaque couche cachée où elle est spécifiée (c'est le paramètre 'rate' dans la fonction), et ceci, afin d'éviter une surconcentration d'informations. Tuer un neurone reviend simplement à fixer son poids à 0. Ainsi, le dropOut a pour objectif de permettre une bonne capacité de généralisation du réseau de neurones construit afin d'améliorer sa capacité de prédiction au vu de nouvelles images. Cette a déjà fait ses preuves dans le monde du deep learning. Les proportions de neurones à tuer varient généralement entre 20% et 50%.

Nous utilisons la classe **ImageDataGenerator** de la librairie **keras** qui est très fluide et efficace pous ces types d'opérations. La première opération à effectuer est la réduction des données d'entrées dans l'intervalle [0,1]. Dans le traitement des images par les modèles de réseaux de neurones, cette pratique est rès fréquente. Puisque les pixels contenus dans les images sont en couleur (RGB), leur valeur selon chaque composante varie entre 0 et 255. Il suffit de diviser les données par le facteur *255* pour revenir à l'échelle [0,1].

### 2. Phase d'entrainement

Un paramètre important lors de l'entrainement de notre réseau est le nombre d'étapes à faire par epoch. On dit qu'un réseau de neurone a effectué **1 epoch** lorsqu'il a utilisé exactement **1 fois** la totalité des observations (ou images) disponible dans la base d'entrainement. Puisque le batch_size spécifie le nombre d'images utilisé par étape d'entrainement, nous en déduisons qu'il faut $\frac{N}{BS}$ étapes pour atteindre **1 epoch**, où $N$ est la taille de l'échantillon d'entrainement et $BS$ le batch_size. Nous allons spécifier ce paramètre maintenant, on le nommera **nb_steps_per_epoch**.

## Q4. Chargement du modèle préentrainé

Comme énoncé précédemment, le temps d'attente est non négligeagle. Pour résoudre ce problème, nous avons alors recours au transfert learning, qui consiste à utiliser des réseaux de neurones préalablement entraînés sur d'autres bases données images qui sont immenses. En l'occurrence, nous utilisons un réseau déjà entrainé sur les 10 catégories d'images présentes dans CIFAR-10. Nous allons charger ce modèle et l'utiliser directement.

```python
pretrained_model = load_model('./pretrained_model.hdf5')
```

Au vu du modèle chargé, nous jugeons adéquat d'enlever les deux dernières couches (Dense et Activation) pour les remplacer par celles qui conviennent mieux.

La méthode **pop()** de **kears** pour enlever la dernière couche d'un réseau de eurones ne fonctionne pas sur la configuration **tensorflow** de notre PC. Nous avons dû définir une fonction **drop_last** qui le fait manuellement.

```python
def drop_last(model):
    name_last_layer = str(model.layers[-1])

    model2 = Sequential()
    for layer in model.layers:
        if str(layer) != name_last_layer:
            model2.add(layer)
    return model2
```

Nous allons finalement définir une nouvelle fonction **transfert_learnig_CNN** qui nous permettra d'appliquer d transfert learning à partir d'un réseau préentrainé donné en paramètre. Il faut ensuite spécifier le nombre de dernière couches à enlever et la dimension de sortie. Nous utilisons pluôt la classe **Model** de **keras** au lieu de **Sequentiel** pour couper les dernières couches et rajouter le nôtre.

```python
def transfert_learnig_CNN(pretrained_model, nb_last_layers_to_drop, output_dim):
    
    #couper autant de fois la dernière couche demandée
    for _ in range(nb_last_layers_to_drop):
        pretrained_model = drop_last(pretrained_model)
    
    #Empilement des modèles
    x = Dense(units=50)(pretrained_model.get_output_at(-1))
    x = Dense(units=output_dim)(x)
    o = Activation(activation="softmax")(x)
    
    opt = rmsprop(learning_rate=0.0001, decay=1e-6)
    model2 = Model(inputs=pretrained_model.input, outputs=[o])
    
    model2.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    return model2
```

```python
model3 = transfert_learnig_CNN(pretrained_model=pretrained_model, nb_last_layers_to_drop=2, output_dim=2)
```

```python
model3.summary()
```

Nous ajoutons d'abord une couche dense de 50 neurones et ensuite une autre de 2 neurones pour la classification des deux classes. Il est important de noter que seuls les poids entrainables sont ceux que nous avons ajouté dans le réseau, de ce fait les poids déjà entrainé par le réseau de neurones chargés restent fixe pendant l'entrainement.
